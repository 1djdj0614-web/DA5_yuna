```python
#모델링 시작
X = cc_df.drop(columns=['is_fraud'])
y = cc_df['is_fraud']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
print("Train fraud rate:", y_train.mean())
print("Test fraud rate:", y_test.mean())
```

```python
#원핫인코딩
X_train_enc = pd.get_dummies(X_train, drop_first=True)
X_test_enc = pd.get_dummies(X_test, drop_first=True)
```

```python
#train과 test 컬럼 맞추기
X_train_enc, X_test_enc = X_train_enc.align(
    X_test_enc,
    join='left',
    axis=1,
    fill_value=0
)
```

```python
#trans_date_trans_time 제거
X_train2 = X_train.drop(columns=['trans_date_trans_time'])
X_test2  = X_test.drop(columns=['trans_date_trans_time'])

X_train_enc = pd.get_dummies(X_train2, drop_first=True)
X_test_enc  = pd.get_dummies(X_test2, drop_first=True)

X_train_enc, X_test_enc = X_train_enc.align(X_test_enc, join='left', axis=1, fill_value=0)
```

```python
#랜덤포레스트 모델 생성
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train_enc, y_train)

```

```python
rf.n_features_in_
```

```python
from sklearn.metrics import roc_auc_score, classification_report, average_precision_score, confusion_matrix

y_prob = rf.predict_proba(X_test_enc)[:, 1]
y_pred = rf.predict(X_test_enc)

print("ROC-AUC:", roc_auc_score(y_test, y_prob))
print("PR-AUC:", average_precision_score(y_test, y_prob))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=4))
```

```python
# 사기인데, 정상으로 오판하는게 높음. 임계값을 조정해야 함
```

```python
#임계값 비교
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

y_prob = rf.predict_proba(X_test_enc)[:, 1]

for thr in [0.5, 0.3, 0.2, 0.1, 0.05]:
    y_pred_thr = (y_prob >= thr).astype(int)
    p = precision_score(y_test, y_pred_thr, zero_division=0)
    r = recall_score(y_test, y_pred_thr)
    f = f1_score(y_test, y_pred_thr)
    cm = confusion_matrix(y_test, y_pred_thr)
    print(f"thr={thr:0.2f} | precision={p:0.4f} recall={r:0.4f} f1={f:0.4f} | FP={cm[0,1]} FN={cm[1,0]} TP={cm[1,1]}")
```

```python
# 임계값이 0.2일 때 fn이 83, fp가 77.
# 임계값이 0.1일 때 fn이 65로 줄어들지만 fp가 239로 폭팔.
# 사기를 잘 잡는 것도 중요하지만, 정상 건을 오판하면 고객센터에 고객 불만이 증가해서 비용도 만만치 않을 거 같음. 
# 임계값 0.2 정도가 적절해 보임.
```

```python
#거리 변수 만들어야겠다
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (km)

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    return R * c
```

```python
# 거리 칼럼 추가
cc_df['distance_km'] = haversine(
    cc_df['lat'],
    cc_df['long'],
    cc_df['merch_lat'],
    cc_df['merch_long']
)
```

```python
#Train 기준 업종 평균 계산
cat_mean = X_train.groupby('category')['amt'].mean()
cat_std  = X_train.groupby('category')['amt'].std()

cat_mean
```

```python
#Train에 적용 (Z-score 방식)
X_train['amt_z_cat'] = (
    (X_train['amt'] - X_train['category'].map(cat_mean)) /
    X_train['category'].map(cat_std)
)
```

```python
#Test에도 적용
X_test['amt_z_cat'] = (
    (X_test['amt'] - X_test['category'].map(cat_mean)) /
    X_test['category'].map(cat_std)
)
```

```python
X_train.assign(is_fraud=y_train).groupby('is_fraud')['amt_z_cat'].mean()
```

```python
X_train_enc = pd.get_dummies(X_train.drop(columns=['trans_date_trans_time']), drop_first=True)
X_test_enc  = pd.get_dummies(X_test.drop(columns=['trans_date_trans_time']), drop_first=True)

X_train_enc, X_test_enc = X_train_enc.align(
    X_test_enc,
    join='left',
    axis=1,
    fill_value=0
)
```

```python
rf2 = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf2.fit(X_train_enc, y_train)
```

```python
y_prob2 = rf2.predict_proba(X_test_enc)[:, 1]

print("ROC-AUC:", roc_auc_score(y_test, y_prob2))
print("PR-AUC:", average_precision_score(y_test, y_prob2))
```

```python
y_pred_02 = (y_prob2 >= 0.2).astype(int)

from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred_02))
print(classification_report(y_test, y_pred_02, digits=4))
```
