```python
#cars title을 brand와 model로 분리
cars = cars_raw.copy()
cars['brand'] = cars['title'].str.split().str[0].str.lower().str.strip()
cars['model'] = cars['title'].str.split().str[1:].str.join(' ')

brand = brand.rename(columns={'title': 'brand'})
brand['brand'] = brand['brand'].str.lower().str.strip()

df = cars.merge(brand[['brand', 'country']], on='brand', how='left')
df.head(10)
```

```python
import numpy as np

df['Engine'] = df['Engine'].astype(str)
engine_num = df['Engine'].str.extract(r'(\d+(\.\d+)?)')[0]
df['engine_L'] = pd.to_numeric(engine_num, errors='coerce')

df.engine_L
```

```python
df['Emission Class'] = df['Emission Class'].astype(str).str.lower().str.strip()
euro_num = df['Emission Class'].str.extract(r'euro\s*([0-9]+)')[0]
df['euro_class'] = pd.to_numeric(euro_num, errors='coerce')

df.euro_class
```

```python
df = df.drop(columns=['Engine', 'Emission Class'])

df
```
